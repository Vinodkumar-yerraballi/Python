```python
'''
Write a Python program that takes a sentence as input and reverses each word in the sentence while keeping 
the word order the same. Example: 
Input: Hello World Python
Output: olleH dlroW nohtyP

'''
# user input
word="Hello World Python"
# create a list to store the values
a=[]
# split the word and store in the list format
revers=word.split() # when we split the sentence we get this output in the step['Hello', 'World', 'Python']
# loop the sentence
for i in revers:
    a.append(i[::-1]) # output ['olleH', 'dlroW', 'nohtyP']
result=" ".join(a)
print(result)


def reverse_sentence(word):
    reverse=word.split()
    result=[i[::-1] for i in reverse] # output ['olleH', 'dlroW', 'nohtyP']
    output=" ".join(result)
    return output
print(reverse_sentence(word))
```

```python
'''
imple elevator system in Python for a building with 25 floors (1 to 25).
The elevator should be represented on the console as [] next to the floor it is currently on. 
The floors should be displayed in descending order (25 at the top, 1 at the bottom). 
The elevator should move step by step between floors (not jump directly). 
After each move, the screen should refresh to show the elevator’s new position. 
The program should ask the user for input (target floor). 
The elevator should then move to that floor.
Invalid floor inputs (outside 1–25) should be rejected. 
The program should keep running until the user enters 0 to exit

'''
def elevator_program(floor):
    for i in range(25,0,-1):
        if i==floor:
            print(print(f"{i:2} []"))
        else:
            print(f"{i:2}")
            
def main_process():
    current_floor=1
    elevator_program(current_floor)
    
    while True:
        try:
            user_input=int(input('user input'))
        except ValueError:
            print('Invalid floor')
            continue
        if user_input==0:
            print('The program exit')
            break
        if user_input < 1 or user_input >25:
            print('invalid floors')
        while current_floor !=user_input:
            if current_floor < user_input:
                current_floor +=1
            else:
                current_floor -=1
            elevator_program(current_floor)
main_process()
            
        
            
```

```python
import os
import time

# Function to display the elevator position
def display_elevator(current_floor):
    os.system('cls' if os.name == 'nt' else 'clear')  # Clear screen for fresh view
    for floor in range(25, 0, -1):  # Show from 25 down to 1
        if floor == current_floor:
            print(f"{floor:2}  []")  # Elevator position
        else:
            print(f"{floor:2}")

# Main program
def elevator_system():
    current_floor = 1  # Start at floor 1
    display_elevator(current_floor)

    while True:
        try:
            target_floor = int(input("\nEnter target floor (1-25) or 0 to exit: "))
        except ValueError:
            print("Invalid input! Please enter a number.")
            continue

        if target_floor == 0:
            print("Exiting elevator system. Goodbye!")
            break
        elif target_floor < 1 or target_floor > 25:
            print("Invalid floor! Please choose between 1 and 25.")
            continue

        # Move step by step
        while current_floor != target_floor:
            if current_floor < target_floor:
                current_floor += 1
            else:
                current_floor -= 1
            display_elevator(current_floor)
            time.sleep(0.3)  # Delay for animation

# Run the program
elevator_system()
```
